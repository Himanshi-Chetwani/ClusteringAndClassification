```python
#Import Statements
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create data frame
col_names = ['ID', 'Title','tconst','Avg_Rating','Num_Votes','Region','Language','isOriginalTitle','TitleType','isAdult',
                'RuntimeMins','Genre']
data = pd.read_csv("C:/Users/anubh/Desktop/2188/sample_data.txt",header=None, names=col_names, sep='\t')
data['Opinion'] = [1 if x >= 5.0 else 0 for x in data['Avg_Rating']]
data.head()
```

```python
#Modify runtime minutes to replace null values
data.RuntimeMins = data.RuntimeMins.replace(to_replace = r'\N', value=0)
data.RuntimeMins = data.RuntimeMins.astype(int)
```

```python
#Histogram of average rating
plt.figure(figsize=(10,6))
sns.distplot(data.Avg_Rating)
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Histogram for Average Rating')
plt.show()
```

```python
#Histogram of number of votes
plt.figure(figsize=(10,6))
plt.plot(data.Num_Votes)
plt.xlabel('Number of Votes')
plt.ylabel('Frequency')
plt.title('Histogram for Number of Votes')
plt.show()
```

```python
#Histogram of run time minutes
plt.figure(figsize=(10,6))
sns.distplot(data.RuntimeMins)
plt.xlabel('Runtime Minutes')
plt.ylabel('Frequency')
plt.title('Histogram for Runtime Minutes')
plt.show()
```

```python
#Pairwise plot of average rating vs number of votes
sns.regplot(x=data.Avg_Rating, y=data.Num_Votes, fit_reg=True)
plt.xlabel('Average Rating')
plt.ylabel('Number of Votes')
plt.title('Plot of average rating vs number of votes')
plt.show()
```

```python
#Pairwise plot of average rating vs run time minutes
sns.regplot(x=data.Avg_Rating, y=data.RuntimeMins, fit_reg=True)
plt.xlabel('Average Rating')
plt.ylabel('Runtime Minutes')
plt.title('Plot of average rating vs runtime minutes')
plt.show()
```
